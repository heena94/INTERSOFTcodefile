##### Plotting from an IPython notebook
Plotting interactively within an IPython notebook can be done with the %matplotlib command, and works in a similar way to the IPython shell. In the IPython notebook, you also have the option of embedding graphics directly in the notebook, with two possible options:

%matplotlib notebook will lead to interactive plots embedded within the notebook
%matplotlib inline will lead to static images of your plot embedded in the notebook
For this book, we will generally opt for %matplotlib inline:

## Plotting in Pandas

```python
graph = pd.Series(np.random.normal(size=64))
print(graph)
graph.plot()
```

```python
graph.cumsum().plot(grid=True)
```

```python
var = pd.DataFrame({'normal': np.random.normal(size=25), 
                       'gamma': np.random.gamma(2, size=25), 
                       'poisson': np.random.poisson(size=25)})
print(var)

var.cumsum().plot()
```

```python
variables = pd.DataFrame({'normal': np.random.normal(2,size=10), 
                       'gamma': np.random.gamma(1, size=10), 
                       'poisson': np.random.poisson(3,size=10)})
print(variables)

variables.cumsum(0).plot()
```

```python
variables = pd.DataFrame({'normal': np.random.normal(size=10), 
                       'gamma': np.random.gamma(1, size=10), 
                       'poisson': np.random.poisson(size=10)})
print(variables)

variables.cumsum(0).plot()
```

```python
variables.cumsum(1).plot(subplots=True)
```

```python
variables.cumsum().plot(secondary_y='normal')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(4, 5))
for i,var in enumerate(['normal','gamma','poisson']):
    variables[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```

## Simple Line Plots

```python
fig = plt.figure()
ax = plt.axes()

s = np.linspace(0, 30, 50)
print(s)
ax.plot(s, np.tan(s));
```

```python
plt.plot(s, np.tan(s))
```

```python
plt.plot(s, np.tan(s - 0), color='red')        
plt.plot(s, np.tan(s - 1), color='0.55')           
plt.plot(s, np.tan(s - 2), color='0.23')     
```

```python
plt.plot(s, np.tan(s - 0), linestyle='-')        
plt.plot(s, np.tan(s - 1), linestyle='--')           
plt.plot(s, np.tan(s - 2), linestyle='-.')  
```

## Adjusting the Plot: Axes Limits

```python
plt.plot(s, np.tan(s - 0))

plt.xlim(-3, 20)
plt.ylim(-2, 10);
```

## Labeling Plots

```python
plt.plot(s, np.sin(s))
plt.title("sin Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

## Visualizing Errors

### Errorbars

```python
a = np.linspace(0, 10, 100)
dy = 0.1
y = np.cos(a) + dy * np.random.randn(100)

plt.errorbar(a, y, yerr=dy, fmt='-',color='0.22');
```
